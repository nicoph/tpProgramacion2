### Valor nulo
Muchas veces, los datos que tenemos presentan valores nulos, NaNs. 

```python
df.loc[10, 'a'] = 2
df
```

```python
# veamos que pasa cuando recalculamos la columna d
df['d'] = df['a'] + df['b']
df
```

Cuando queremos sumar dos celdas en la que una hay un valor nulo el resultado es otro valor nulo. Esto tiene sentido ya que pandas no sabe como sumar un valor nulo con un valor no nulo. 

Todos estamos tentados de reemplazar ese valor nulo por un cero, no? Pero c√≥mo sabemos si est√° bien reemplazarlo por un cero o por un -1 o por un 10? La respuesta es que depende del contexto del trabajo que estemos haciendo. 

Supongamos que en este trabajo tiene sentido reemplazar el valor nulo por un 0. Tenemos varias formas de hacer eso:
- reemplazar el NaN por un zero inplace
- reemplazar el NaN por un zero solo para la operaci√≥n en cuesti√≥n

```python
df['a'] + df['b'].fillna(0)
```

```python
df['a'].add(df['b'], fill_value = 0)
```

```python
df['b'].fillna(0, inplace = True)
df['b']
```

### Duplicated, Sort, Unique
Supongamos que no queremos tener valores rerpetidos de listings. 
1. Vamos a fijarnos si tenemos valores repetidos. Eso lo podemos hacer con el m√©todo `.duplicated`

```python
data[data['listing_url'].duplicated(keep = 'first')]
```

Por lo que vemos arriba, en este dataset no hay listings repetidos. 

Se ve que lo limpiaron bien üßπüßπüßπüßπüßπüßπ

Supongamos solo queremos tener un listing por `host_id` para poder calcular `host_total_listings_count` medio de esta ciudad

```python
data[~(data['host_id'].duplicated(keep = 'last'))]['host_total_listings_count'].mean()
```

```python
# Chequeamos que la longitud del dataframe resultante sea la misma que la cantidad de host_ids unicos
len(data[~(data['host_id'].duplicated(keep = 'last'))]), len(data['host_id'].unique())
```

```python
data[~(data['host_id'].duplicated(keep = 'last'))]['host_total_listings_count'].median()
```

Ahora supongamos que queremos quedarnos con un √∫nico registro por host pero quedarnos solo con aquel registro que tiene la mayor cantidad de `bedrooms`. Entonces, primero, ordenamos los datos y luego los filtramos con `duplicated`

```python
data.sort_values(by = 'bedrooms', inplace = True, ascending = False)
```

```python
# luego, miremos cuales son os host_ids que estan duplicados para tener una idea
data[data['host_id'].duplicated(keep = False)]['host_id']
```

```python
# miramos un ejemplo en particular
data[data['host_id'] == 4271294][['bedrooms']]
```

```python
data[(~(data['host_id'].duplicated(keep = 'first')))&(data['host_id']==4271294)][['bedrooms']]
```

## Tipos de datos
Como ya se habr√°n dado cuenta, los datos dentro de las Series y los DFs tiene un tipo de dato ü§∑
Los datos puede ser de cualquier tipo y los podemos pasar de un tipo de dato a otro. 

Abajo vemos que el id es un n√∫mero.  Muchas veces, resulta √∫til guardar los ids como strings. Esto lo podemos hacer con el m√©todo `astype`

```python
data['id'].head(5).tolist()
```

```python
data['id'] = data['id'].astype('str')
```

```python
data['id'].head(5).tolist()
```

Tambi√©n es √∫til definir el tipo de dato cuando leemos el archivo para evitar p√©rdidas de informaci√≥n. Por ejemplo, supongamos que tenemos un id que empieza con ceros, 000034. 
Cuando pandas lea ese dato, si no le decimos que es una string, lo va a leer como un n√∫mero y vamos a perder los 0000.

Para definir el tipo de dato podemos hacer lo siguiente:

```python
data = pd.read_csv('listings.csv.gz', dtype = {'id':'str'})
```

```python
data.id.tolist()[:10]
```
